# Classify different data sets

### Basic includes

```python
# Using pandas to load the csv file
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt

from keras import models 
from keras import layers 
from keras import callbacks
from keras.utils import to_categorical

# reuters and fashin mnist data set from keras
from keras.datasets import reuters
from keras.datasets import fashion_mnist

# needed to preprocess text
from keras.preprocessing.text import Tokenizer
```

### Classify the Fashion Mnist

---

```python
(fashion_train_data, fashion_train_labels), (fashion_test_data, fashion_test_labels) = fashion_mnist.load_data()

print(fashion_train_data.shape)

test_index = 10

plt.title("Label: " + str(fashion_train_labels[test_index]))
plt.imshow(fashion_train_data[test_index], cmap="gray")
```

#### TO DO: Preprocess the data

1. Normalize the input data set
2. Perform one hot encoding
3. Create a train, test, and validation set

```python
fashion_class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

fashion_train_data = fashion_train_data.reshape(60000, 28 * 28)
fashion_train_data = fashion_train_data.astype('float32') / 255
fashion_test_data = fashion_test_data.reshape(10000, 28 * 28)
fashion_test_data = fashion_test_data.astype('float32') / 255

fashion_train_labels = to_categorical(fashion_train_labels)
fashion_test_labels = to_categorical(fashion_test_labels)

fashion_validation_data = fashion_train_data[:5000]
fashion_validation_labels = fashion_train_labels[:5000]

fashion_x_data = fashion_train_data[5000:]
fashion_y_data = fashion_train_labels[5000:]
```

#### TO DO: Define and train a network, then plot the accuracy of the training, validation, and testing

1. Use a validation set
2. Propose and train a network
3. Print the history of the training
4. Evaluate with a test set

```python
fashion_NN_model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(28 * 28,)),
    layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(10, activation='softmax')
])

early_stop = callbacks.EarlyStopping(monitor='val_loss', patience=5)

fashion_NN_model.compile(optimizer='adam', 
              loss='categorical_crossentropy',
              metrics=['accuracy'])

fashion_NN_model.summary()

fashion_history = fashion_NN_model.fit(fashion_x_data, fashion_y_data, epochs=10, validation_data=(fashion_validation_data, fashion_validation_labels), callbacks=[early_stop], verbose=2)

fashion_results = fashion_NN_model.evaluate(fashion_test_data, fashion_test_labels)

print('Fashion Test accuracy:', fashion_results)

# This dictionary stores the validation and accuracy of the model throughout the epochs
fashion_history_dict = fashion_history.history
print(fashion_history_dict.keys())

# The history values are split in different lists for ease of plotting
fashion_acc = fashion_history_dict['acc']
fashion_val_acc = fashion_history_dict['val_acc']
fashion_loss = fashion_history_dict['loss']
fashion_val_loss = fashion_history_dict['val_loss']

fashion_epochs = range(1, len(fashion_acc) + 1)

# Plot of the validation and training loss

# "bo" is for "blue dot"
plt.plot(fashion_epochs, fashion_loss, 'bo', label='Training loss')
# b is for "solid blue line"
plt.plot(fashion_epochs, fashion_val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()

# Plot of the validation and train accuracy

plt.clf()   # clear figure

plt.plot(fashion_epochs, fashion_acc, 'bo', label='Training acc')
plt.plot(fashion_epochs, fashion_val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

#### Fashion_comments
For this model, the tendency was to overfit and then adjust the parameters to improve the precision.



## Classifying newswires

---

Build a network to classify Reuters newswires into 46 different mutually-exclusive topics.

### Load and review the data

```python
(reuters_train_data, reuters_train_labels), (reuters_test_data, reuters_test_labels) = reuters.load_data(num_words=10000)

print(reuters_train_data.shape)
print(reuters_train_labels.shape)
print(reuters_train_data[0])
print(reuters_train_labels[0])

print(set(reuters_train_labels))
```

Load the word index to decode the train data.

```python
word_index = reuters.get_word_index()

reverse_index = dict([(value+3, key) for (key, value) in word_index.items()])

reverse_index[0] = "<PAD>"
reverse_index[1] = "<START>"
reverse_index[2] = "<UNKNOWN>"  # unknown
reverse_index[3] = "<UNUSED>"

decoded_review = ' '.join([reverse_index.get(i,'?') for i in reuters_train_data[0]])

print(decoded_review)
```

#### TO DO: Preprocess the data

1. Normalize the input data set
2. Perform one hot encoding
3. Create a train, test, and validation set

```python
tokenizer = Tokenizer(num_words=10000)
reuters_train_data_token = tokenizer.sequences_to_matrix(reuters_train_data, mode='binary')
reuters_test_data_token = tokenizer.sequences_to_matrix(reuters_test_data, mode='binary')

# One-hot encoding the output
reuters_train_labels = to_categorical(reuters_train_labels, 46)
reuters_test_labels = to_categorical(reuters_test_labels, 46)

# Creating a validation set with the first 10000 reviews
reuters_validation_data = reuters_train_data_token[:1000]
reuters_validation_labels = reuters_train_labels[:1000]

# Creating the input set
reuters_x_data = reuters_train_data_token[1000:]
reuters_y_data = reuters_train_labels[1000:]
```

#### TO DO: Define and train a network, then plot the accuracy of the training, validation, and testing

1. Use a validation set
2. Propose and train a network
3. Print the history of the training
4. Evaluate with a test set

```python
reuters_NN_model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(10000,)),
    layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(46, activation='softmax')
    ])

reuters_NN_model.compile(optimizer='rmsprop', 
              loss='categorical_crossentropy',
              metrics=['accuracy'])

reuters_NN_model.summary()
    
reuters_history = reuters_NN_model.fit(reuters_x_data, reuters_y_data, epochs=10, batch_size=512, validation_data=(reuters_validation_data,reuters_validation_labels))
    
reuters_results = reuters_NN_model.evaluate(reuters_test_data_token, reuters_test_labels)

print('Reuters Test accuracy:', reuters_results)

# This dictionary stores the validation and accuracy of the model throughout the epochs
reuters_history_dict = reuters_history.history
print(reuters_history_dict.keys())

# The history values are split in different lists for ease of plotting
reuters_acc = reuters_history_dict['acc']
reuters_val_acc = reuters_history_dict['val_acc']
reuters_loss = reuters_history_dict['loss']
reuters_val_loss = reuters_history_dict['val_loss']

reuters_epochs = range(1, len(reuters_acc) + 1)

# Plot of the validation and training loss

# "bo" is for "blue dot"
plt.plot(reuters_epochs, reuters_loss, 'bo', label='Training loss')
# b is for "solid blue line"
plt.plot(reuters_epochs, reuters_val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()

# Plot of the validation and train accuracy

plt.clf()   # clear figure

plt.plot(reuters_epochs, reuters_acc, 'bo', label='Training acc')
plt.plot(reuters_epochs, reuters_val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

Using the preprocessing seen in classes and then tuning the parameters of the model, i couldn't get past 80 of accuracy.
Even the testing tended to overfit.


## Predicting Student Admissions

---

Predict student admissions based on three pieces of data:

- GRE Scores
- GPA Scores
- Class rank

### Load and visualize the data

```python
student_data = pd.read_csv("student_data.csv")
print(student_data)
```

Plot of the GRE and the GPA from the data.

```python
X = np.array(student_data[["gre","gpa"]])
y = np.array(student_data["admit"])
admitted = X[np.argwhere(y==1)]
rejected = X[np.argwhere(y==0)]
plt.scatter([s[0][0] for s in rejected], [s[0][1] for s in rejected], s = 25, color = 'red', edgecolor = 'k')
plt.scatter([s[0][0] for s in admitted], [s[0][1] for s in admitted], s = 25, color = 'cyan', edgecolor = 'k')
plt.xlabel('Test (GRE)')
plt.ylabel('Grades (GPA)')

plt.show()
```

Plot of the data by class rank.

```python
f, plots = plt.subplots(2, 2, figsize=(20,10))
plots = [plot for sublist in plots for plot in sublist]

for idx, plot in enumerate(plots):
    data_rank = student_data[student_data["rank"]==idx+1]
    plot.set_title("Rank " + str(idx+1))
    X = np.array(data_rank[["gre","gpa"]])
    y = np.array(data_rank["admit"])
    admitted = X[np.argwhere(y==1)]
    rejected = X[np.argwhere(y==0)]
    plot.scatter([s[0][0] for s in rejected], [s[0][1] for s in rejected], s = 25, color = 'red', edgecolor = 'k')
    plot.scatter([s[0][0] for s in admitted], [s[0][1] for s in admitted], s = 25, color = 'cyan', edgecolor = 'k')
    plot.set_xlabel('Test (GRE)')
    plot.set_ylabel('Grades (GPA)')
    
```

#### TO DO: Preprocess the data

1. Normalize the input data set
2. Perform one hot encoding
3. Create a train, test, and validation set

```python
student_data = student_data.fillna(0)

normalized_student_data = pd.get_dummies(student_data, columns=['rank'])
normalized_student_data["gre"] = normalized_student_data["gre"] / 800
normalized_student_data["gpa"] = normalized_student_data["gpa"] / 4

np.random.shuffle(normalized_student_data.values)

student_x = np.array(normalized_student_data)[:,1:]
student_x = student_x.astype('float32')
student_y = to_categorical(student_data["admit"])

# Creating a validation set 
student_validation_data = student_x[:100]
student_validation_labels = student_y[:100]

# Creating the input set
student_x_data = student_x[100:]
student_y_data = student_y[100:]
```

#### TO DO: Define and train a network, then plot the accuracy of the training, validation, and testing

1. Use a validation set
2. Propose and train a network
3. Print the history of the training
4. Evaluate with a test set

```python
student_NN_model = models.Sequential([
    layers.Dense(128, activation='relu', kernel_initializer='random_uniform', input_shape=(7,)),
    layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(32, activation='relu'),
    layers.Dense(2, activation='softmax')
    ])

student_NN_model.compile(optimizer='rmsprop', 
              loss='categorical_crossentropy',
              metrics=['accuracy'])

student_NN_model.summary()
    
student_history = student_NN_model.fit(student_x_data, student_y_data, epochs=15, batch_size=100, validation_data=(student_validation_data,student_validation_labels))
    
student_results = student_NN_model.evaluate(student_x[80:200], student_y[80:200])

print('student Test accuracy:', student_results)

# This dictionary stores the validation and accuracy of the model throughout the epochs
student_history_dict = student_history.history
print(student_history_dict.keys())

# The history values are split in different lists for ease of plotting
student_acc = student_history_dict['acc']
student_val_acc = student_history_dict['val_acc']
student_loss = student_history_dict['loss']
student_val_loss = student_history_dict['val_loss']

student_epochs = range(1, len(student_acc) + 1)

# Plot of the validation and training loss

# "bo" is for "blue dot"
plt.plot(student_epochs, student_loss, 'bo', label='Training loss')
# b is for "solid blue line"
plt.plot(student_epochs, student_val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()

# Plot of the validation and train accuracy

plt.clf()   # clear figure

plt.plot(student_epochs, student_acc, 'bo', label='Training acc')
plt.plot(student_epochs, student_val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
